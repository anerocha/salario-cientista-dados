## 2\. Análise Exploratória de Dados (EDA)

### **2.1.Distribuição de Salários**

Vamos começar com uma análise geral da distribuição de salários de cientistas de dados.



```python
plt.figure(figsize=(10,6))
sns.histplot(df['salary_in_usd'], kde=True, color='blue', bins=30)
plt.title('Distribuição de Salários de Cientistas de Dados (em USD)')
plt.xlabel('Salário (em USD)')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

### **2.2.Salário por Nível de Experiência**

Agora, vamos analisar como o nível de experiência afeta o salário. Consideraremos as categorias *Entry-level, Mid-level, Senior-level, Executive-level.*

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='experience_level', y='salary_in_usd', palette='Set2')
plt.title('Salário por Nível de Experiência')
plt.xlabel('Nível de Experiência')
plt.ylabel('Salário (em USD)')
plt.show()
```

### **2.3.Salário por Tipo de Emprego**



Vamos também verificar se o tipo de emprego (Full-time, Part-time, Freelance, etc.) tem impacto no salário.

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='employment_type', y='salary_in_usd', palette='Set1')
plt.title('Salário por Tipo de Emprego')
plt.xlabel('Tipo de Emprego')
plt.ylabel('Salário (em USD)')
plt.show()
```

### **2.4.Salário por Localização**



Como o local de trabalho também pode influenciar o salário, vamos analisar os salários por país.

```python
country_salary = df.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
country_salary.plot(kind='bar', color='orange')
plt.title('Top 10 Países com Maiores Salários Médios de Cientistas de Dados')
plt.xlabel('País')
plt.ylabel('Salário Médio (em USD)')
plt.xticks(rotation=45)
plt.show()
```

### **2.5.Salário por Tamanho da Empresa**


Também vamos analisar como o tamanho da empresa (small, medium, large) impacta o salário dos cientistas de dados.

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='company_size', y='salary_in_usd', palette='coolwarm')
plt.title('Salário por Tamanho da Empresa')
plt.xlabel('Tamanho da Empresa')
plt.ylabel('Salário (em USD)')
plt.show()
```

### **2.6.Top 10 Países com Maiores Salários**


```python
country_salary = df.groupby('employee_residence')['salary_in_usd'].mean()

top_10_countries_highest_salary = country_salary.sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_10_countries_highest_salary.plot(kind='bar', color='green')
plt.title('Top 10 Países com Maiores Salários Médios de Cientistas de Dados')
plt.xlabel('País')
plt.ylabel('Salário Médio (em USD)')
plt.xticks(rotation=45)
plt.show()
```

### **2.7.Top 10 Países com Menores Salários**


```python
country_salary = df.groupby('employee_residence')['salary_in_usd'].mean()

top_10_countries_lowest_salary = country_salary.sort_values(ascending=True).head(10)

plt.figure(figsize=(12,6))
top_10_countries_lowest_salary.plot(kind='bar', color='red')
plt.title('Top 10 Países com Menores Salários Médios de Cientistas de Dados')
plt.xlabel('País')
plt.ylabel('Salário Médio (em USD)')
plt.xticks(rotation=45)
plt.show()
```

## 3\. Análises Focadas no Brasil

### **3.1.Dados de Cientistas de Dados no Brasil**


Primeiro, filtramos os dados para mostrar apenas os cientistas de dados que residem no Brasil.

```python
brazil_data = df[df['employee_residence'] == 'BR']

brazil_data.head()
```

### **3.2.Salário Médio no Brasil**


vamos calcular o salário médio de cientistas de dados que residem no Brasil.

```python
avg_salary_brazil = brazil_data['salary_in_usd'].mean()

print(f'O salário médio de cientistas de dados no Brasil é: ${avg_salary_brazil:.2f} USD')
```

### **3.3.Distribuição de Salários no Brasil**


```python
plt.figure(figsize=(10,6))
sns.histplot(brazil_data['salary_in_usd'], kde=True, color='blue', bins=20)
plt.title('Distribuição dos Salários de Cientistas de Dados no Brasil')
plt.xlabel('Salário (em USD)')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

### **3.4.Comparação do Brasil com Outros Países**


Agora, vamos comparar o salário médio de cientistas de dados no Brasil com o salário médio de outros países.

```python
country_salary = df.groupby('employee_residence')['salary_in_usd'].mean()

brazil_salary_comparison = pd.concat([country_salary.loc[['BR']], country_salary.drop('BR').sort_values(ascending=False).head(10)])

plt.figure(figsize=(12,6))
brazil_salary_comparison.plot(kind='bar', color=['blue', 'green'])
plt.title('Comparação de Salário Médio entre o Brasil e Outros Países')
plt.xlabel('País')
plt.ylabel('Salário Médio (em USD)')
plt.xticks(rotation=45)
plt.show()
```

## 4\. Conclusões e Insights

*Após a análise, podemos tirar algumas conclusões importantes:*





1.   **Distribuição Global de Salários:**  *A maioria dos salários de cientistas de dados está concentrada em uma faixa intermediária, com valores variando entre S50.000 a S120.000 USD anuais, enquanto poucos profissionais alcançam salários muito altos ou baixos.*


2.   **Impacto do Nível de Experiência:** *Profissionais mais experientes, como Senior e Executive-level, possuem salários significativamente mais altos, superando S150.000 USD anuais, enquanto os de nível Entry-level recebem entre S50.000 a S70.000 USD.*


3. **Diferenças por Localização:** *Países como EUA, Suíça e Austrália oferecem os salários mais altos, superando S120.000 USD. O Brasil, por outro lado, tem uma média de S20.000 a S35.000 USD anuais.*


4. **Comparação com o Brasil:** *O Brasil apresenta um salário médio de cientistas de dados inferior aos países líderes, mas ainda assim oferece uma remuneração atrativa, especialmente quando comparado com o custo de vida.*
